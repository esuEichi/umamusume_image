```python
import cv2
import numpy as np
import shutil
```

```python
filePath = "star"
```

```python
import os
# フォルダ内の全ファイル名をリスト化
files = os.listdir(filePath)

# リストの長さ（ファイル数）を取得
count = len(files) - 1
```

```python
baff = 10
yellow = 20

filenames = []

for num in range(count):
    filename = filePath + '/' + str(num) + '.jpg'
    img = cv2.imread(filename)
    _, width, _= img.shape[:3]
    left = int(width/8)
    img = img[:,left:]
    
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    lower = (yellow - baff, 110, 200)
    upper = (yellow + baff, 200, 255)
    star_img = cv2.inRange(hsv, lower, upper)
    element = cv2.getStructuringElement(cv2.MORPH_RECT, (2, 2))
    # star_img = cv2.morphologyEx(star_img, cv2.MORPH_CLOSE, element)
    star_img = cv2.dilate(star_img, element, iterations=2)
    _, contours, _ = cv2.findContours(star_img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)

    out_filename = 'results/results/' + str(num)+ '_star_' + str(len(contours)) + '.jpg'
    
    cv2.imwrite(out_filename, img)
    filenames.append(out_filename)
```
